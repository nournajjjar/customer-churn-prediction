```python
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display, HTML

# Ajouter un bouton interactif pour masquer/afficher le code dans UNE seule cellule
display(HTML('''
<style>
    .toggle-button { 
        text-align: center; 
        font-family: sans-serif; 
        font-size: 14px; 
        color: #444; 
        cursor: pointer; 
        margin: 10px 0;
    }
    .toggle-button:hover { color: blue; }
</style>
<script>
    function toggleCurrentCode() {
        var currentCell = this.closest('.cell');
        var codeInput = currentCell.querySelector('.input');
        if (codeInput.style.display === 'none') {
            codeInput.style.display = '';
        } else {
            codeInput.style.display = 'none';
        }
    }
</script>
<div class="toggle-button" onclick="toggleCurrentCode.call(this)">
    &#9650; Show hidden code &#9660;
</div>
<hr style="border: none; border-top: 1px solid lightgray;">
'''))

# Calculate outliers info
outliers_info = {}

for column in fusion.select_dtypes(include=['float64', 'int64']).columns:
    Q1 = fusion[column].quantile(0.25)
    Q3 = fusion[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Count outliers
    outliers_count = ((fusion[column] < lower_bound) | (fusion[column] > upper_bound)).sum()
    outliers_info[column] = outliers_count  # Only store the count of outliers

# Sorting columns by the count of outliers
outliers_info = {k: v for k, v in sorted(outliers_info.items(), key=lambda item: item[1], reverse=True)}

# Display the outliers_info dictionary
print(outliers_info)

# Increase figure size for clarity
plt.figure(figsize=(15, 10))  # Adjust figure size as needed

# Create the boxplot with custom colors
sns.boxplot(data=fusion, palette=['#6497bf', '#ffb9b9', '#d1b3ff'])

# Add a title with larger font size
plt.title('Boxplot des variables pour détecter les valeurs aberrantes', fontsize=18, weight='bold')

# Rotate x-axis labels for better readability
plt.xticks(rotation=90, fontsize=12)

# Add grid for better visual aid
plt.grid(True, linestyle='--', alpha=0.6)

# Optionally adjust the y-axis range (modify values as per your data)
plt.ylim(-100, 550)

# Ensure proper layout adjustment for visibility
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
from scipy.cluster.hierarchy import dendrogram, linkage
from matplotlib import pyplot as plt
from matplotlib.colors import LinearSegmentedColormap
display(HTML('''
<style>
    .toggle-button { 
        text-align: center; 
        font-family: sans-serif; 
        font-size: 14px; 
        color: #444; 
        cursor: pointer; 
        margin: 10px 0;
    }
    .toggle-button:hover { color: blue; }
</style>
<script>
    function toggleCurrentCode() {
        var currentCell = this.closest('.cell');
        var codeInput = currentCell.querySelector('.input');
        if (codeInput.style.display === 'none') {
            codeInput.style.display = '';
        } else {
            codeInput.style.display = 'none';
        }
    }
</script>
<div class="toggle-button" onclick="toggleCurrentCode.call(this)">
    &#9650; Show hidden code &#9660;
</div>
<hr style="border: none; border-top: 1px solid lightgray;">
'''))
# Assuming the function cah_profiling is already defined
def cah_profiling(data, selected_features, n_clusters):
    # Normalize the data
    scaler = StandardScaler()
    data_scaled = scaler.fit_transform(data[selected_features])

    # Create linkage matrix and dendrogram
    linkage_matrix = linkage(data_scaled, method='ward')
    plt.figure(figsize=(10, 7))

    # Customize dendrogram with color_threshold and custom colors
    dendrogram(linkage_matrix, color_threshold=20)

    # Customizing colors in the dendrogram after it's created
    # Custom color logic
    plt.title("Dendrogram for CAH")
    plt.xlabel("Samples")
    plt.ylabel("Euclidean Distance")
    plt.show()

    # Agglomerative Clustering
    cah_model = AgglomerativeClustering(n_clusters=n_clusters, linkage='ward')
    data['cluster_result'] = cah_model.fit_predict(data_scaled) + 1

    # Summarize Mean Profiles for Each Cluster
    df_profile_overall = pd.DataFrame()
    df_profile_overall['Overall'] = data.describe().loc[['mean']].T

    df_cluster_summary = data.groupby('cluster_result').describe().T.reset_index().rename(
        columns={'level_0': 'Column Name', 'level_1': 'Metrics'})
    df_cluster_summary = df_cluster_summary[df_cluster_summary['Metrics'] == 'mean'].set_index('Column Name')

    # Combine Overall and Cluster Summaries
    df_profile = df_cluster_summary.join(df_profile_overall).reset_index()
    return df_profile

# Define custom color gradient (light blue to light red)
custom_colors = ["#ADD8E6", "#ffb9b9", "#DDA0DD" , "#87CEEB" ,"#FFC0CB","#4B0082"]  # Light Blue to Light Red
custom_cmap = LinearSegmentedColormap.from_list("custom_cmap", custom_colors)

# Main workflow example
if __name__ == "__main__":
    # Assuming the dataset is already imported as 'fusion'
    data = fusion  # Use the fusion dataset directly

    # Defining features
    selected_features = fusion.columns.tolist()  # Use all columns from fusion dataset

    # Perform CAH Profiling
    n_clusters = 3
    profiling_results = cah_profiling(data, selected_features, n_clusters)

    # Display profiling results with custom color gradient
    print("\n--- Profiling Results ---\n")
    print(profiling_results.style.background_gradient(cmap=custom_cmap))
```
